```python
import pandas as pd
```

```python
def make_table(taxon_id, number_rows=10):
    df = pd.read_csv(f'./results/{taxon_id}/{taxon_id}_network_analysis.csv')
    df.fillna('-', inplace=True)
    df.rename(
        columns={
            'name': 'gene name',
            'MCODE::Clusters (1)': 'MCODE cluster',
            'AverageShortestPathLength': 'average shortest path length',
            'ClusteringCoefficient': 'cluster coefficient',
            'ClosenessCentrality':'closeness centrality',
        },
        inplace=True
    )
    df.sort_values(by='average shortest path length', ascending=True, inplace=True)
    df = df.round(2)
    return df[['gene name', 'MCODE cluster', 'average shortest path length', 'cluster coefficient', 'closeness centrality']].head(number_rows)
```

- Enterococcus faecalis (strain ATCC 700802 / V583) / taxon id 226185

```python
taxon_id = 226185
make_table(taxon_id)
```
